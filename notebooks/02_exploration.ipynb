```python
from collections import Counter

def flatten_lists(series):
    out = []
    for x in series:
        if isinstance(x, list):
            out.extend(x)
    return out

top_genres = Counter(flatten_lists(catalog["genres"])).most_common(20)
top_countries = Counter(flatten_lists(catalog["countries"])).most_common(20)
top_genres[:10], top_countries[:10]
```
