```python
from collections import Counter
import pandas as pd
import numpy as np

def ensure_list(x):
    # already a collection?
    if isinstance(x, (list, tuple, set)):
        return list(x)

    # numpy array (your case)
    if isinstance(x, np.ndarray):
        return x.tolist()

    # missing
    if x is None:
        return []
    if isinstance(x, float) and pd.isna(x):
        return []

    # string like "A, B, C"
    if isinstance(x, str):
        return [t.strip() for t in x.split(",") if t.strip()]

    return []

genres = catalog["genres"].apply(ensure_list)
countries = catalog["countries"].apply(ensure_list)

top_genres = Counter(g for row in genres for g in row).most_common(20)
top_countries = Counter(c for row in countries for c in row).most_common(20)
top_genres[:10], top_countries[:10]
```
