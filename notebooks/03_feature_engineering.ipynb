```python
from pathlib import Path
import pandas as pd

from src.feature_engineering import (
    build_core_feature_matrix, CoreFeatureConfig,
    build_cast_feature_matrix, CastFeatureConfig
)

CATALOG_PATH = Path("data/processed/catalog.parquet")

catalog = pd.read_parquet(CATALOG_PATH)
catalog.shape
```

```python
cfg = CoreFeatureConfig(top_genres=30, top_countries=25)
X_core, core_meta = build_core_feature_matrix(catalog, cfg)

X_core.shape, X_core.columns[:10]
```

```python
OUT_CORE = Path("data/processed/features_core.parquet")
OUT_CORE.parent.mkdir(parents=True, exist_ok=True)
X_core.to_parquet(OUT_CORE, index=True)

OUT_CORE
```

```python
cast_cfg = CastFeatureConfig(top_actors=200)
X_cast, cast_meta = build_cast_feature_matrix(catalog, cast_cfg)

X_cast.shape
```

```python
OUT_CAST = Path("data/processed/features_cast.parquet")
X_cast.to_parquet(OUT_CAST, index=True)
OUT_CAST
```
